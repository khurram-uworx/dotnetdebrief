# ✨ AI Assisted Engineering

👉 Prompting: A new Engineering Skill

Prompting is like writing a spec:

* Bad: “optimize this code”
* Better: “optimize for readability and O(n) time; avoid extra memory; keep API unchanged”

Prompting Concepts:

* Context matters more than clever wording
* Iteration beats one-shot prompting
* Constrain outputs (format, style, tests)

🧠 Mental Model:

* Writing good tickets
* Writing clean function signatures
