## 🛠️ Tools

```
User Goal → Planner → Tool Use + Memory → Execution → Reflection → Final Output
```

- Tool Use Protocol (Function Calling Layer)
    - Agents need a clean, structured way to interact with tools (APIs, databases, code execution)
    - Extensions in VS Code coming with their own Tools
- Memory Protocol (Short & Long Term)
    - Agents are useless if they forget everything between steps
    - retrieval > raw context stuffing
- Planning & Decomposition
    - Modern agents shouldn’t just react—they should plan multi-step actions
    - Task decomposition (break big goals into steps)
    - Tree/graph planning (branching strategies)
    - Replanning loops (adjust after failure)
- Reflection / Self Critique Loop
- Execution + Sandbox Protocol

# ✨ AI Assisted Engineering

👉 Prompting: A new Engineering Skill

Prompting is like writing a spec:

* Bad: “optimize this code”
* Better: “optimize for readability and O(n) time; avoid extra memory; keep API unchanged”

Prompting Concepts:

* Context matters more than clever wording
* Iteration beats one-shot prompting
* Constrain outputs (format, style, tests)

🧠 Mental Model:

* Writing good tickets
* Writing clean function signatures
